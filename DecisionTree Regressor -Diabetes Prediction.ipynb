```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

## data : 10 features
## target : output - has a continuous value because its regression problem

## independent features
      - age     age in years
      - sex
      - bmi     body mass index
      - bp      average blood pressure
      - s1      tc, total serum cholesterol
      - s2      ldl, low-density lipoproteins
      - s3      hdl, high-density lipoproteins
      - s4      tch, total cholesterol / HDL
      - s5      ltg, possibly log of serum triglycerides level
      - s6      glu, blood sugar level

```python
X=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
## independent feature
X
```

```python
# dependent feature
y=dataset['target']
```

```python
y
```

```python
X.head()
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
## check correlation
X_train.corr()
```

```python
## heatmap for showing corr
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
## Apply Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
treeRegressor=DecisionTreeRegressor()
treeRegressor.fit(X_train,y_train)
```

```python
## Hyper paraneter tuning
params={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter': ['best','random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
    }
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
regressor=DecisionTreeRegressor()
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
## check with respect to other params in new model
newmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
newmodel.fit(X_train,y_train)
```

```python
## visualize decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(newmodel,filled=True)
```

```python
y_pred_new=newmodel.predict(X_test)
```

```python
print(r2_score(y_test,y_pred_new))
print(mean_absolute_error(y_test,y_pred_new))
print(mean_squared_error(y_test,y_pred_new))
```
