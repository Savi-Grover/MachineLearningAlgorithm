## Cleaning Data

```python
## taking first set by header
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset.csv', header=1)
```

```python
df.head()
```

```python
df.info()
```

```python
## data cleaning check missing values
df.isnull().sum()
```

```python
## retrieving rows with null value 
df[df.isnull().any(axis=1)]

## we see header 2 row has null as its set 2 header
## only row has no classes value
```

```python
## also we see upto row 122 - we have data for bijae region
## After 122 - we have data for Sidi-Bel Abbes Region
```

```python
## add a new column , where label =0 for bijae region and 1 for sidi-bel-abbes region

df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.info()
```

```python
df['Region'] .dtype
```

```python
## converting 1, 0 float to int

df[['Region']]=df[['Region']].astype(int)
```

```python
## removing 1 and 122 row and reset index
df=df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## we see column headers - day month temp are in 122 row so drop
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

## Monthly Fire Analysis on df

```python
## providing labels to region in df 

df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis on df ( Sidi-Bel Abbes Region)

dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('no of fires', weight='bold')
plt.xlabel('months', weight='bold')
plt.title('Fire Analysis of Sidi-Bel Abbes Region', weight='bold')
```

```python
## Monthly Fire Analysis on df ( Bejai Region)

dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('no of fires', weight='bold')
plt.xlabel('months', weight='bold')
plt.title('Fire Analysis of Bejai Region', weight='bold')
```
