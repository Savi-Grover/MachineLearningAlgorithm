```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
## create database ( not using any external csv)
from sklearn.datasets import make_classification
```

```python
# producing input and ouput features which 1000 datapoints, 10 features, 3 outputs, we also need informatives
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
## output - will be 0, 1 ,2
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
## implement Logistic Regression - multi_class will have 3 values = auto ; ovr; 
from sklearn.linear_model import LogisticRegression
Logistic=LogisticRegression(multi_class='ovr')
Logistic.fit(X_train,y_train)
```

```python
y_pred=Logistic.predict(X_test)
y_pred
```

```python
##Peformance metrics

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
print(score)
# confusion matric will 2*2 since we have binary problem
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
